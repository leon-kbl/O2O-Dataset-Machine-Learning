### 浏览数据，将数据的折扣情况统计出来，并绘图（自由发挥）

```python
%matplotlib inline
# %matplotlib inline   #魔法方法，内嵌图，有这个就可以省掉plt.show，还有就是不能加注释，不然这个魔法方法不起作用。
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import log_loss, roc_auc_score, auc, roc_curve
import lightgbm as lgb
```

```python
data=df['Discount_rate'].value_counts()
data=data.sort_index()
fig,ax=plt.subplots(figsize=(18,4))
ax= sns.barplot(data.index,data.values,alpha=0.8)
ax.xaxis.set_tick_params(rotation=60)
plt.title("Discount_rate Distribution", fontsize=20)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Discount_rate', fontsize=12)
rects = ax.patches
for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(int(height)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
Y=[]
df = df[df['Discount_rate'] == df['Discount_rate']]
#统计折扣为折扣率的个数
Y.append(df[df['Discount_rate'].str.contains('.')]['Discount_rate'].count())
#统计折扣为空值的个数
Nan_num = df['Discount_rate'].count()
Y.append(200000 - Nan_num)
#统计折扣为满减的个数
Y.append(df[df['Discount_rate'].str.contains(':')]['Discount_rate'].count())
Y
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
#绘制柱状图
X = ['折扣率', '无折扣', '满减']
plt.figure(figsize=(10, 8))
x=range(len(X))
for i, y in enumerate(Y):
    plt.text(i, y, '%s' % y, ha='center', va='bottom')
plt.xticks(x, X)
plt.bar(x,Y,width=0.5,color=['r', 'b', 'g'])
```

```python
fig,ax=plt.subplots(figsize=(10,8))
ax= sns.barplot(X,Y,alpha=0.8)
rects = ax.patches
for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(int(height)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
plt.title("折扣类型统计", fontsize=20)
plt.xlabel('折扣方式', fontsize=12)
plt.ylabel('折扣数量', fontsize=12)
```

```python
#绘制饼状图
plt.pie(Y, colors=['r','b','g'], labels=X)
plt.axis("equal")
plt.title("折扣类型统计")
plt.legend()
```

### 使用random forest进行标准数据建模，查看各特征权重，并作图（自由发挥）

```python
df['Discount_rate']=df['Discount_rate'].fillna(0)
df['Discount_rate_new']=df['Discount_rate'].apply(lambda x:x if ':' not in str(x) else ((float(x.split(':')[0])-float(x.split(':')[1]))/float(x.split(':')[0])))
df['Discount_rate_new']=df['Discount_rate_new'].apply(lambda x:round(float(x),2))
from sklearn.model_selection import train_test_split
df['label']=df['Date'].apply(lambda x:1 if x==x else 0)
y=df['label']
from sklearn.model_selection import train_test_split
X=df
df['label']=df['Date'].apply(lambda x:1 if x==x else 0)
y=df['label']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
selected_features=['Merchant_id','Coupon_id','Distance','Discount_rate_new']
```

```python
df.head()
```

```python
df.info()
```

```python
rf = RandomForestClassifier(max_depth = 10)
rf.fit(X_train.loc[:,selected_features].fillna(0), y_train)
y_rfpred =  [x[1]for x in rf.predict_proba(X_test.loc[:,selected_features].fillna(0))]
```

```python
auc_rf = roc_auc_score(y_test,y_rfpred)
print('用Random Forest 进行预测得到的auc：',auc_rf)
```

```python
rf.feature_importances_
```

```python

# Create a pd.Series of features importances
y=rf.feature_importances_.tolist()
x=selected_features
df=pd.DataFrame({'important':y,'name':x})  #从新组一个DataFrame,为了排序
sorted_importances_rf = df.sort_values(by='important')
sorted_importances_rf.set_index('name',inplace=True)  # 让name列变为索引列
# # Make a horizontal bar plot
sorted_importances_rf.plot(kind='barh', color='lightgreen')
```

### 使用LightGBM进行标准数据建模并调参

原生态lightgbm


```python
lgb_train = lgb.Dataset(X_train.loc[:,selected_features].fillna(0), y_train)  
param = {
    'num_leaves':150, 
    'objective':'binary',
    'max_depth':10,
    'learning_rate':.1,
    'max_bin':100}
param['metric'] = ['auc', 'binary_logloss']
num_round = 100
lgbm = lgb.train(param,lgb_train,num_round)
ypred2 = lgbm.predict(X_test.loc[:,selected_features].fillna(0))
auc_lgbm = roc_auc_score(y_test,ypred2)
print('用LightGBM 进行预测得到的auc：',auc_lgbm)
```

##### sklearn   LGBMClassifier

```python
import lightgbm as lgb
lgb_params = {
    'learning_rate':0.1,
    'max_bin':150,
    'num_leaves':32,    
    'max_depth':9,  
    'objective':'binary',
    'n_estimators':300
}
model=lgb.LGBMClassifier(**lgb_params)
model.fit(X_train.loc[:,selected_features].fillna(0),y_train)
# 预测
#输出预测结果
test_pre2 = model.predict(X_test.loc[:,selected_features].fillna(0))
auc_lgbm = roc_auc_score(y_test,ypred2)
print('用LGBMClassifier  进行预测得到的auc：',auc_lgbm)
```

##### sklearn LGBMRegressor

```python
import lightgbm as lgb
from lightgbm import LGBMRegressor
lgb_params={
    'learning_rate':0.1,
    'max_bin':150,
    'num_leaves':32,    
    'max_depth':9,  
    'objective':'binary',
    'n_estimators':300
}
gbm = LGBMRegressor(**lgb_params)
gbm.fit(X_train.loc[:,selected_features], y_train, eval_set=[(X_test.loc[:,selected_features], y_test)], eval_metric='l1', early_stopping_rounds=100)
```

```python
y_gbm = gbm.predict(X_test.loc[:,selected_features], num_iteration=gbm.best_iteration_)

print('用LGBMRegressor  进行预测得到的auc：',roc_auc_score(y_test, y_gbm))
```

### 按照此模型预测第200,000-210,000行的AUC score。要求 AUC>0.7

```python
df=pd.read_csv('ccf_offline_stage1_train-Copy1.csv',skiprows=range(1,200001))
df['Discount_rate']=df['Discount_rate'].fillna(0)
df['Discount_rate_new']=df['Discount_rate'].apply(lambda x:x if ':' not in str(x) else ((float(x.split(':')[0])-float(x.split(':')[1]))/float(x.split(':')[0])))
df['Discount_rate_new']=df['Discount_rate_new'].apply(lambda x:round(float(x),2))
from sklearn.model_selection import train_test_split
df['label']=df['Date'].apply(lambda x:1 if x==x else 0)
y=df['label']
from sklearn.model_selection import train_test_split
X=df
df['label']=df['Date'].apply(lambda x:1 if x==x else 0)
y=df['label']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
selected_features=['Merchant_id','Coupon_id','Distance','Discount_rate_new']
```

```python
rf = RandomForestClassifier(max_depth = 10)
rf.fit(X_train.loc[:,selected_features].fillna(0), y_train)
y_rfpred =  [x[1]for x in rf.predict_proba(X_test.loc[:,selected_features].fillna(0))]
auc_rf = roc_auc_score(y_test,y_rfpred)
print('用Random Forest 进行预测得到的auc：',auc_rf)
```

```python
lgb_train = lgb.Dataset(X_train.loc[:,selected_features].fillna(0), y_train)  
param = {
    'num_leaves':150, 
    'objective':'binary',
    'max_depth':10,
    'learning_rate':.1,
    'max_bin':100}
param['metric'] = ['auc', 'binary_logloss']
num_round = 100
lgbm = lgb.train(param,lgb_train,num_round)
ypred2 = lgbm.predict(X_test.loc[:,selected_features].fillna(0))
auc_lgbm = roc_auc_score(y_test,ypred2)
print('用LightGBM 进行预测得到的auc：',auc_lgbm)
```

2.将训练数据(ccf_offline_stage1_train.csv)的**全部数据**读入Pandas dataframe

```python
import pandas as pd
#df = pd.read_csv('./ccf_offline_stage1_train.csv', nrows = 200000)
df = pd.read_csv('./ccf_offline_stage1_train.csv')
df.head()
```

3.使用Matrix factorization对转化率建模，并对比该结果与GBDT模型结果

```python
df['Discount_rate']=df['Discount_rate'].fillna(0)
df['Discount_rate_new']=df['Discount_rate'].apply(lambda x:x if ':' not in str(x) else ((float(x.split(':')[0])-float(x.split(':')[1]))/float(x.split(':')[0])))
df['Discount_rate_new']=df['Discount_rate_new'].apply(lambda x:round(float(x),2))
from sklearn.model_selection import train_test_split
df['label']=df['Date'].apply(lambda x:1 if x==x else 0)
y=df['label']
X=df
df['label']=df['Date'].apply(lambda x:1 if x==x else 0)
y=df['label']
selected_features=['Merchant_id','Coupon_id','Discount_rate_new','Distance']
x_train,x_test,y_train,y_test=train_test_split(X.loc[:,selected_features].fillna(0),y,test_size=0.3,random_state=0)
```

```python
import lightgbm as lgb
train_data=lgb.Dataset(x_train,y_train)
eval_data=lgb.Dataset(x_test,y_test)
```

```python
lgb_params = {
    'boosting_type': 'gbdt',  
    'objective': 'binary',
    'metric': 'auc', 
    'learning_rate':0.01,
     'max_depth':4 }
clf = lgb.train(lgb_params,train_data)
                
eval_pre=clf.predict(x_test)
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,eval_pre)
```

```python
User_id_df=df[['User_id']].fillna(0).drop_duplicates()
User_id_df
User_id_df['User_new_id']=User_id_df.apply(lambda x:str(int(x['User_id'])),axis=1)
User_id_df['idx']=np.arange(len(User_id_df))
User_id_dict=User_id_df[['User_new_id','idx']].set_index('User_new_id').iloc[:,0].to_dict()
```

```python
df['Merchant_id']=df['Merchant_id'].fillna(0)
df['Coupon_id']=df['Coupon_id'].fillna(0)
df['User_id']=df['User_id'].fillna(0)
```

```python
df['Mer_cou_id']=df['Merchant_id'].map(str)+'_'+df['Coupon_id'].map(str)
df['Mer_cou_id']=df['Mer_cou_id'].apply(lambda x:int(x.split('.')[0]))
df.head()
```

```python
train_df,eval_df=train_test_split(df,random_state=0)
```

```python
# %pip install findspark
# import findspark
# findspark.init()
import  pyspark
```

```python
sc=pyspark.SparkContext()
spark=pyspark.sql.SparkSession(sc)
```

```python
train_df.head()
```

```python
train_sdf=spark.createDataFrame(train_df[['Mer_cou_id','User_id','label']])
test_sdf=spark.createDataFrame(eval_df[['Mer_cou_id','User_id','label']])
```

```python
from pyspark.ml.recommendation import ALS
from pyspark.sql import functions as f
```

```python
train_sdf.show()
```

```python
als=ALS(rank=10,maxIter=5,seed=0,userCol='User_id',itemCol='Mer_cou_id',ratingCol='label')
model=als.fit(train_sdf)
pred_sdf=model.transform(test_sdf)
pred_sdf.limit(5).toPandas()
```

```python
pred_sdf.fillna(0).groupby('label').agg(f.mean('prediction')).toPandas()
```

```python
mf_df=pred_sdf.fillna(0.5).toPandas()
mf_df
```

```python
roc_auc_score(mf_df.label,mf_df.prediction)
```
